# Anomalias

```python
# Ler os dados para cada cenário e modelo - IPSL
df1 = pd.read_csv("CMIP.IPSL.IPSL-CM6A-LR.historical.csv", parse_dates=['time'], usecols=['time', 'media_thetao', 'media_so'], index_col='time')
df2 = pd.read_csv("ScenarioMIP.IPSL.IPSL-CM6A-LR.ssp585.csv", parse_dates=['time'], usecols=['time', 'media_thetao', 'media_so'], index_col='time')
df_ipsl = pd.concat([df1, df2])

# Ler os dados para cada cenário e modelo - MIROC
df3 = pd.read_csv("CMIP.MIROC.MIROC6.historical.csv", parse_dates=['time'], usecols=['time', 'media_thetao', 'media_so'], index_col='time')
df4 = pd.read_csv("ScenarioMIP.MIROC.MIROC6.ssp585.csv", parse_dates=['time'], usecols=['time', 'media_thetao', 'media_so'], index_col='time')
df_miroc = pd.concat([df3, df4])

# Ler os dados para cada cenário e modelo - CESM
df5 = pd.read_csv("CMIP.NCAR.CESM2.historical.csv", parse_dates=['time'], usecols=['time', 'media_thetao', 'media_so'], index_col='time')
df6 = pd.read_csv("ScenarioMIP.NCAR.CESM2.ssp585.csv", parse_dates=['time'], usecols=['time', 'media_thetao', 'media_so'], index_col='time')
df_cesm = pd.concat([df5, df6])

mean_ipsl = df_ipsl.loc['1980':'2010'].mean()
anom_ipsl = df_ipsl - mean_ipsl

mean_miroc = df_miroc.loc['1980':'2010'].mean()
anom_miroc = df_miroc - mean_miroc

mean_cesm = df_cesm.loc['1980' : '2010'].mean()
anom_cesm = df_cesm - mean_cesm

# Calcule a média entre os DataFrames dos três modelos
df_mean = pd.DataFrame({
    'media_thetao': (df_cesm['media_thetao'] + df_miroc['media_thetao'] + df_ipsl['media_thetao']) / 3,
    'media_so': (df_cesm['media_so'] + df_miroc['media_so'] + df_ipsl['media_so']) / 3
}, index=df_cesm.index)  # Use qualquer um dos índices dos DataFrames

mean_ensemble = df_mean.loc['1980' : '2010'].mean()
anom_ensemble = df_mean - mean_ensemble
```

### Anomalias Salinidade

```python
plt.figure(figsize=(10,4))
# Plot dos dados para cada modelo com legendas
plt.plot(anom_cesm.index, anom_cesm.media_so, label='CESM2', color="blue")
plt.plot(anom_miroc.index, anom_miroc.media_so, label='MIROC6', color="gold")
plt.plot(anom_ipsl.index, anom_ipsl.media_so, label='IPSL-CM6A-LR', color="crimson")

# Adicionar legendas ao gráfico
plt.legend()
plt.title("Anomalia de Salinidade")
# Adicione rótulos aos eixos x e y
plt.ylabel('Anomalias da Média de Salinidade')
plt.savefig("Salinity_anomaly.png")
# Exiba o gráfico
plt.show()
```

```python
def std_T(anom):
    std_percentage = np.linspace(0, 0.4, len(anom.index))
    std_1 = anom.media_thetao * std_percentage
    return std_1

std_cesm = std_T(anom_cesm)
std_miroc = std_T(anom_miroc)
std_ipsl = std_T(anom_ipsl)

plt.figure(figsize=(10,4))
# Plot dos dados para cada modelo com legendas
plt.plot(anom_cesm.index, anom_cesm.media_thetao, label='CESM2', color="blue")
plt.plot(anom_miroc.index, anom_miroc.media_thetao, label='MIROC6', color='crimson')
plt.plot(anom_ipsl.index, anom_ipsl.media_thetao, label='IPSL-CM6A-LR', color='gold')

plt.fill_between(anom_cesm.index, anom_cesm.media_thetao - std_cesm, anom_cesm.media_thetao + std_cesm, color='blue', alpha=0.3)
plt.fill_between(anom_miroc.index, anom_miroc.media_thetao - std_miroc, anom_miroc.media_thetao + std_miroc, color='crimson', alpha=0.3)
plt.fill_between(anom_ipsl.index, anom_ipsl.media_thetao - std_ipsl, anom_ipsl.media_thetao + std_ipsl, color='gold', alpha=0.3)


# Adicionar legendas ao gráfico
plt.legend()

# Adicione rótulos aos eixos x e y
plt.ylabel('Anomalias da Média de Temperatura(°C)')
plt.title("Anomalias de Temperatura")
plt.grid()
# Exiba o gráfico
plt.savefig("Anomalias_Temperatura.png")
plt.show()
```
